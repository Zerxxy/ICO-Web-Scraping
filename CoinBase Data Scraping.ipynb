```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
```

```python
baseUrl = 'https://coinmarketcap.com'
page = urlopen(baseUrl)
soup = BeautifulSoup(page, 'html.parser')
```

```python
table = soup.find_all('tbody')
table
```

```python
trueTable = table[0]
print(trueTable.prettify())
```

```python
rows = trueTable.find_all('tr')
row1 = rows[0]
print(row1.prettify())
```

```python
links= row1.find_all("a", href=True)
second_page = baseUrl+links[0]['href']
second_page
```

```python
page2 = urlopen(second_page)
soup2 = BeautifulSoup(page2, 'html.parser')

table2 = soup2.find_all('ul', class_="list-unstyled")
table2[0].find_all('li')[0].find_all("a", href=True)[0]['href']
```

```python
links[1].get_text()
```

```python
coinNames = []
coinLinks = []

for index in range(len(rows)):
    row = rows[index]
    links = row.find_all("a", href=True)
    next_page = baseUrl+links[0]['href']
    
    page3 = urlopen(next_page)
    soup3 = BeautifulSoup(page3, 'html.parser')
    
    table3 = soup3.find_all('ul', class_="list-unstyled")
    
    name = links[1].get_text()
    link = table3[0].find_all('li')[0].find_all("a", href=True)[0]['href']
    
    coinNames.append(name)
    coinLinks.append(link)
    
```

```python
coinDF = pd.DataFrame({
    "name": coinNames,
    "links": coinLinks
})
coinDF
```
