```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
baseUrl = 'https://coinmarketcap.com'
page = urlopen(baseUrl)
soup = BeautifulSoup(page, 'html.parser')
```

```python
table = soup.find_all('tbody')
table
```

```python
trueTable = table[0]
print(trueTable.prettify())
```

```python
rows = trueTable.find_all('tr')
row1 = rows[0]
print(row1.prettify())
```

```python
links= row1.find_all("a", href=True)
second_page = baseUrl+links[0]['href']
second_page
```

```python
page2 = urlopen(second_page)
soup2 = BeautifulSoup(page2, 'html.parser')

```
